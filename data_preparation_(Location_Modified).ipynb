## Data Preparation For 'location_modified.csv'

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pylab as pl
%matplotlib inline
```

```python
customer = pd.read_csv("Data/olist_customers_dataset.csv", na_values=["?"])
geo = pd.read_csv("Data/olist_geolocation_dataset.csv", na_values=["?"])
item = pd.read_csv("Data/olist_order_items_dataset.csv",na_values=["?"])
payment = pd.read_csv("Data/olist_order_payments_dataset.csv",na_values=["?"])
review = pd.read_csv("Data/olist_order_reviews_dataset.csv",na_values=["?"])
order = pd.read_csv("Data/olist_orders_dataset.csv", na_values=["?"])
product = pd.read_csv("Data/olist_products_dataset.csv",na_values=["?"])
seller = pd.read_csv("Data/olist_sellers_dataset.csv",na_values=["?"])
eng = pd.read_csv("Data/product_category_name_translation.csv")

customer = pd.DataFrame(customer)
geo = pd.DataFrame(geo)
item = pd.DataFrame(item)
payment = pd.DataFrame(payment)
review = pd.DataFrame(review)
order = pd.DataFrame(order)
product = pd.DataFrame(product)
seller = pd.DataFrame(seller)
eng = pd.DataFrame(eng)
```

```python
# Join tables of customer and order on customer_id
a = pd.merge(customer, order, on='customer_id')
a.shape
```

```python
a.head()
```

```python
a.columns
```

```python
cols = ['customer_state','order_approved_at', 'order_delivered_carrier_date']
a.drop(cols, axis=1, inplace=True)
a.columns
```

```python
a.rename(columns={'customer_zip_code_prefix': 'zipcode',
                  'order_purchase_timestamp':'timestamp',
                  'order_delivered_customer_date':'delivered_date',
                  'order_estimated_delivery_date':'estimated_date'}, inplace=True)
a.columns
```

```python
np.where(a.duplicated(subset=['customer_id']))
```

```python
geo.columns
```

```python
geo.rename(columns={'geolocation_zip_code_prefix': 'zipcode',
                   'geolocation_lat': 'latitude',
                   'geolocation_lng':'longitude'}, inplace=True)
```

```python
geo.columns
```

```python
cols = ['geolocation_city','geolocation_state']
geo.drop(cols, axis=1, inplace=True)
geo.columns
```

```python
geo.shape
```

```python
a.shape
```

```python
m = a.merge(geo, on='zipcode', how='left')
m.shape
```

```python
m = m.drop_duplicates(subset=['customer_id'])
m.shape
```

```python
m.head()
```

### There are 278 locations that are no geographic data. 

```python
m[m.isnull().any(axis=1)]
```

### Drop NaN value in the table:

```python
# drop the non-geographic location
m = m.dropna() 
m.shape
```

```python
# wrong location(not in Brazil, Portugal)
out = m.loc[(m['latitude'] >= 35.0) & (m['longitude'] >= -10.0)] 
print(out.shape)
out
```

```python
# wrong location(not in Brazil, Argentina)
out = m.loc[(m['latitude'] <= -27.0) & (m['longitude'] <= -58.0)] 
print(out.shape)
out
```

### Drop the location that is not in Brazil:

```python
m = m.drop(m[m.zipcode == 68275].index)
m = m.drop(m[m.zipcode == 83252].index)
m = m.drop(m[m.zipcode == 83810].index)
m = m.drop(m[m.zipcode == 28155].index)
m = m.drop(m.loc[(m['zipcode'] == 98780)].index)

print(m.shape)
m.head()
```

```python
m.info()
```

```python
item.head()
```

```python
item = item.drop('shipping_limit_date', 1)
```

```python
i = m.merge(item, on='order_id', how='left')
i.shape
```

```python
m_table = i.drop_duplicates(subset=['customer_id'])
m_table.shape
```

```python
m_table.head()
```

```python
pd.options.mode.chained_assignment = None
total_order_item_value = m_table.loc[:,'price'] * m_table.loc[:,'order_item_id']
total_freight_value = m_table.loc[:,'freight_value'] * m_table.loc[:,'order_item_id']

m_table.loc[:,'total_value'] = total_order_item_value + total_freight_value
m_table.head()
```
